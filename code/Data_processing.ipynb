# Simple Reorganization and subsetting of data

## GOALS:

### Working on unpublished records.
There are four major subsets (orders) of data in the Insecta class  
1. Lepidoptera
2. Diptera
3. Hymenoptera
4. Coleoptera  

All better listed below in detail:

```python
%%bash
cd /home/kibet/bioinformatics/github/co1_metaanalysis/data/input/input_data/clean_africa
echo -e "A total of `grep ">" enafroCOI_all_data.fasta | wc -l` records"
grep ">" enafroCOI_all_data.fasta | awk 'BEGIN{FS="|"};{print $2}' | sort | uniq -c
```

We divided the data into clusters based on the order in the taxonomic headers: Diptera, Lepidotera, Hymenoptera, Coleoptera
1. **Diptera**

```python
%%bash
cd /home/kibet/bioinformatics/github/co1_metaanalysis/data/input/input_data/
mkdir diptera
cp ./clean_africa/enafroCOI_* ../diptera/
cd ./diptera
. ../../../../code/process_all_input_files.sh
delete_unwanted ./*.fasta << EOF
1
Diptera
2
EOF
echo -e "\nRenaming files...\n"
mv enafroCOI_500to700_data-650to660_undesired.fasta enafroCOI_diptera_500to700_data-650to660.fasta
mv enafroCOI_500to700_data_undesired.fasta enafroCOI_diptera_500to700_data.fasta
mv enafroCOI_650to660_data_undesired.fasta enafroCOI_diptera_650to660_data.fasta
mv enafroCOI_all_data_raw_undesired.fasta enafroCOI_diptera_all_data_raw.fasta
mv enafroCOI_Over499_data_undesired.fasta enafroCOI_diptera_Over499_data.fasta
mv enafroCOI_Over700_data_undesired.fasta enafroCOI_diptera_Over700_data.fasta
mv enafroCOI_Under500_data_undesired.fasta enafroCOI_diptera_Under500_data.fasta
mv enafroCOI_all_data_undesired.fasta enafroCOI_diptera_all_data.fasta

echo -e "\nRemoving residual files"
rm enafroCOI_500to700_data* enafroCOI_650to660_data.fasta enafroCOI_all_data* enafroCOI_Over* enafroCOI_Under500_data.fasta
wc -l *
```
